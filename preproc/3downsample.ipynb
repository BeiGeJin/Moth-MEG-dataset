```python
import os, glob
import mne
```

### Set Parameters

```python
SUBJECT = "A"
SESSION = "Moth2"
BLOCKS = ["01", "02", "03", "04", "05", "06", "07", "08"] if SESSION == "Moth1" or SESSION == "Moth4" else ["01", "02", "03", "04", "05", "06", "07"]
VERSION = "sss_er-4-10-2-2_band-0.1-150_notch-60-120_ecg-ica_eog-ica"
DOWN_FS = 50
LOC_RAW = f"/project_data/volume0/jerryjin/moth_meg/{SESSION}/{VERSION}/{SUBJECT}/"
LOC_SAVE = LOC_RAW + f"downsampled-{DOWN_FS}/"
```

```python
# create directory if not exist
if not os.path.exists(LOC_SAVE):
    os.makedirs(LOC_SAVE)
```

### Run Downsampling

```python
# check all recording files for this session
files = glob.glob(LOC_RAW + "*.fif")
files
```

```python
for BLOCK in BLOCKS:

    # load raw data
    f = glob.glob(LOC_RAW + f"*{SUBJECT}_{SESSION}_{BLOCK}*raw.fif")[0]
    raw = mne.io.read_raw_fif(f, preload=True)  # preload helps accelerate

    # downsample
    raw.resample(DOWN_FS)

    # save downsampled data
    save_f = os.path.basename(f).replace("_raw.fif", f"_ds{DOWN_FS}_raw.fif")
    raw.save(LOC_SAVE + save_f, overwrite=True)
```
